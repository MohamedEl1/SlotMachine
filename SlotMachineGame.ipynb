# Slot machine containing 3 classes and one main function
The three classes are Purse, Column and Slot

- The purse class manages the players account winning and losings. It has a get_balance method that retrieves the current balance, has debit method which takes the bet amount away from the player if he scores 0. The credit method adds the score to the balance.
For the credit i first substract the bet amount from the balance then add the score to the balance and for the debit i substract both the bet amount and the score (0 in this case) instead.


- The column class has a face representing the fruit. it includes a change_face method that would change the fruit on face at random when the pull_handle method on Slot class is invoked


- The slot class has four methods, pull_handle which will call the column class and changes faces, show_slot which prints out the emojis slots, take_bet method which gets the bet amount from the player and passes it on to the score_slot method which uses it to calculate the score based on the faces match.

```python
import emoji
from random import choice
class Purse(): # purse is the balance of how much money you have in the game updates the available money.
    def __init__(self):
        self.balance = 10 
    
    def get_balance(self): 
        return self.balance 
    
        
    def debit(self, score, betted_amount): 
        
        # if the user scores 0 this will take away his bet amount from the balance he has
        
        self.balance = self.balance - betted_amount - int(score)  
        print('you have: ', self.balance)
        
        # credit to add score (winnings) to the balance
    def credit(self, score, betted_amount):
        self.balance = self.balance - betted_amount + int(score)
        print('you have: ', self.balance)
        
     
```

```python
class Column():
    # changes fruit on face on random when the pull handle is invoked
    faces = [emoji.emojize(':red_apple:'),
                 emoji.emojize(':pear:'),
                 emoji.emojize(':tangerine:')]
    def __init__(self):
        self.face = choice(self.faces)

    def change_face(self):
        self.face = choice(self.faces)
```

```python
class Slot(Purse):
    def __init__(self):
        Purse.__init__(self)
            # here we are making instances for the class columns so we can run its methods
        self.column1 = Column()
        self.column2 = Column()
        self.column3 = Column()
    
    def pull_handle(self):  # when you pull the handle the faces will change and calls the coloumns 

        self.column1.change_face()
        self.column2.change_face()
        self.column3.change_face()
        self.show_slot()
    

    def show_slot(self):
        print(self.column1.face, self.column2.face, self.column3.face)

    def take_bet(self, betted_amount): # this method takes the bet and uses it below to calculate the score
            self.betted_amount = betted_amount
        

    def score_slot(self): # the score slot checks the value of three coloumns and depending on that it updates the purse
        if self.column1.face == self.column2.face and self.column1.face == self.column3.face and self.column2.face == self.column3.face:
            # calculates the score based on the above result comparision of faces
            score =  self.betted_amount * 2
            return score
        
        elif self.column1.face == self.column2.face:
            score =  1.5 * self.betted_amount 
            print('you score: ', score)
            return score
        elif self.column1.face == self.column3.face:
            score =  1.5 * self.betted_amount 
            print('you score: ', score)
            return score
        elif self.column2.face == self.column3.face:
            score =  1.5 * self.betted_amount 
            print('you score: ', score)
            return score
        else:
            score = 0 
            print('you score: ', score, '- Thank you for playing')
            return score
            
```

```python
def run_slot_machine(): # Main function of the whole game, it does everything
    print("********Welcome to The Slot Machine Game!********")
    print('Minimum bet is 2, type N to exit')
    machine = Slot()
    playeraccount = Purse()
    print('You have:', playeraccount.get_balance())
    keep_playing = True 
    
    
    while keep_playing == True:
            betted_amount = input('How much do you bet: ')
            if betted_amount == "N":
                keep_playing = False
            else:
                if betted_amount.isdigit():
                    betted_amount = int(betted_amount)
                    if betted_amount < 2 or betted_amount > playeraccount.get_balance():
                        continue
        
                    machine.take_bet(betted_amount)
                    machine.pull_handle()
                    score = machine.score_slot()

        # if the score slot is greater than the betted amount then update the credit purse
                    if score > betted_amount:
                        playeraccount.credit(score, betted_amount)

        # if the score slot is 0 then update then debit purse

                    if score == 0:
                        playeraccount.debit(score, betted_amount)
                    
                    if playeraccount.get_balance() < 2:
                        print('You are leaving with', playeraccount.get_balance())
                        break
                 
```

```python
run_slot_machine()
```
